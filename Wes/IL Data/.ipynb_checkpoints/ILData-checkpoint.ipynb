```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np

data=pd.read_csv('FinalData.csv',sep=',',header=0)

ExpDens=data.ExpDens.tolist()
CalDens=data.AvgDens.tolist()
ExpStdDens=data.ExpStdDens.tolist()
CalStdDens=data.CalStdDens.tolist()
plt.figure(figsize=(15,12))
plt.xlabel('Experimental Specific Density (kg/m^3)')
plt.ylabel('Calculated Specific Density (kg/m^3)')
plt.plot(ExpDens,CalDens,'.',color='black')
plt.errorbar(ExpDens,CalDens,xerr=ExpStdDens,yerr=CalStdDens,ls='none',color='gray')
plt.show()

ExpCpt=data.ExpCpt.tolist()
CalCpt=data.AvgCpt.tolist()
ExpStdCpt=data.ExpStdCpt.tolist()
CalStdCpt=data.CalStdCpt.tolist()
plt.figure(figsize=(15,12))
plt.xlabel('Experimental Specific Heat Capacity (J/K/m^3)')
plt.ylabel('Calculated Specific Heat Capacity (J/K/m^3)')
plt.plot(ExpCpt,CalCpt,'.',color='black')
plt.errorbar(ExpCpt,CalCpt,xerr=ExpStdCpt,yerr=CalStdCpt,ls='none',color='gray')
plt.show()
```
